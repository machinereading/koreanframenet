```python
import json
from pprint import pprint
```

```python
version = '1.1'
fname = '../resource/'+version+'/KFN_annotations.json'
```

```python
with open(fname+'.bak', 'r') as f:
    ori = json.load(f)
```

```python
new = {}
for i in ori:
    anno_id = i
    anno = ori[i]
    
    print('ori')
    pprint(anno)
    for deno in anno['denotations']:
        if deno['role'] == 'TARGET':
            tid = deno['id']
    rels = []
    for deno in anno['denotations']:
        if deno['id'] != tid:
            rel = {}
            rel['subj'] = tid
            rel['pred'] = 'arg'
            rel['obj'] = deno['id']
            rels.append(rel)
    anno['relations'] = rels
    
    print('\nnew')
    pprint(anno)
    
    
    break
```
