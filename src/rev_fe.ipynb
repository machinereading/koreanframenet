```python
with open('./fe_910416mapping_table_.csv','r') as f:
    d = f.readlines()
    
mappings = []

for line in d:
    line = line.strip()
    print(line)
    m = line.split(',')[0], line.split(',')[1], line.split(',')[2]
    
    mappings.append(m)
```

```python
def get_frame(sent):
    for line in sent:
        if line.startswith('#'):
            pass
        else:
            f = line.split('\t')[3]
            if f != '_':
                frame = f
                
    return frame
```

```python
def feconvert(frame, fe):
    result = fe
    for f, ori, new in mappings:
        if frame == f:
            if fe == ori:
                result = new
                
    return result                
```

```python
def rev_fe(ori_fname, fname):
    with open(ori_fname, 'r') as f:
        d = f.readlines()

    sents = []
    sent = []
    with open(fname, 'w') as f:
        for line in d:
            line = line.strip()

            if line != '':
                sent.append(line)
            else:
                sents.append(sent)
                sent = []

    with open(fname, 'w') as f:
        for sent in sents:
            frame = get_frame(sent)

            for line in sent:
                new_line = False

                if line.startswith('#'):
                    new_line = line
                elif line == '':
                    new_line = line
                else:
                    fe = line.split('\t')[-1].split('-')[-1]
                    bio = line.split('\t')[-1].split('-')[0]

                    if fe == 'O':
                        new_line = line

                    else:                
                        if bio == 'S':
                            new_bio = 'B'
                        else:
                            new_bio = bio            
                        new_fe = feconvert(frame, fe)
                        
                        if new_fe == 'O':
                            new_line = line.replace(bio+'-'+fe, new_fe)
                        else:
                            new_line = line.replace(bio+'-'+fe, new_bio+'-'+new_fe)

                f.write(new_line+'\n')
            f.write('\n')
```
